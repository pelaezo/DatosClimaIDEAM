```python
#Cambiamos el formato de la columna Fecha a tipo Fecha(datetime), con la libreria panda

bdcsv['Fecha'] = pd.to_datetime(bdcsv['Fecha'])
```

```python
#Volvemos a revisamos las caracteristicas de cada columna, y observamos que cambio el tipo de dato de 'Fecha'

bdcsv.info()
```

```python
# Extraemos los datos de Fecha y Valor
Var_Clima = bdcsv[['Fecha','Valor']]
```

```python
#Se obtiene la menor fecha y la mayor
MIN = Var_Clima['Fecha'].min()
MAX = Var_Clima['Fecha'].max()
print(MIN, MAX)
```

```python
import datetime
```

```python
YEAR = Var_Clima['Fecha'].dt.year
MINYEAR = YEAR.min()
MAXYEAR = YEAR.max()
YEARS = (MAXYEAR - MINYEAR)
print(MINYEAR, MAXYEAR, YEARS)
LIMYEAR = datetime.date((MINYEAR+YEARS),12,1)
print(LIMYEAR)
```

```python
YEAR = Var_Clima['Fecha'].dt.year
MINYEAR = YEAR.min()
MAXYEAR = YEAR.max()
YEARS = (MAXYEAR - MINYEAR)
print(MINYEAR, MAXYEAR, YEARS)
LIMAXYEAR = datetime.date((MINYEAR+YEARS),12,1)
print(LIMAXYEAR)
LIMINYEAR = datetime.date((MAXYEAR-YEARS),1,1)
print(LIMINYEAR)
```

```python
# Se verifica si hay datos mensuales faltantes
Var_Clima.groupby(Var_Clima['Fecha'].dt.month_name()).count()
```

```python
# Se identifican cuales son los meses faltantes

# Crear un rango de fechas deseado
rangocomparar = pd.date_range(start=LIMINYEAR, end=LIMAXYEAR,  freq= 'MS')

#print(rangocomparar)

# Encontrar las fechas faltantes
comparar = rangocomparar.difference(Var_Clima['Fecha'])

#convertimos el dato a DataFrame
comparar1 = pd.DataFrame({'Fecha': comparar})

print(comparar1)
```

```python
#Verificamos el Tipo de dato
result2=type(comparar1)
print(result2)
```

```python
# Unimos la db del IDEAM con la bd Faltantes
union_bd = pd.concat([Var_Clima, comparar1])
print(union_bd)
```

```python
union_bd = union_bd.sort_values(by='Fecha')
print(union_bd)
```

```python
union_bd
```

```python
# Filtrar las filas correspondientes al mes de Noviembre
df_november = union_bd[union_bd['Fecha'].dt.month_name() == 'November']

print(df_november)
```

```python
# Creamos la tabla de valores por mes y año

df_January = union_bd[union_bd['Fecha'].dt.month_name() == 'January']
df_February = union_bd[union_bd['Fecha'].dt.month_name() == 'February']
df_March = union_bd[union_bd['Fecha'].dt.month_name() == 'March']
df_April = union_bd[union_bd['Fecha'].dt.month_name() == 'April']
df_May = union_bd[union_bd['Fecha'].dt.month_name() == 'May']
df_June = union_bd[union_bd['Fecha'].dt.month_name() == 'June']
df_July = union_bd[union_bd['Fecha'].dt.month_name() == 'July']
df_August = union_bd[union_bd['Fecha'].dt.month_name() == 'August']
df_September = union_bd[union_bd['Fecha'].dt.month_name() == 'September']
df_October = union_bd[union_bd['Fecha'].dt.month_name() == 'October']
df_November = union_bd[union_bd['Fecha'].dt.month_name() == 'November']
df_December = union_bd[union_bd['Fecha'].dt.month_name() == 'December']

YEAR9 = df_January['Fecha'].dt.year
```

```python
# Creamos un dataFrame
tb_mes_ano = pd.DataFrame({
    'Año': YEAR9.tolist(), # Ubica la información del año
    'ENE': df_January['Valor'].tolist(), # Extrae el 'Valor' de la columna
    'FEB': df_February['Valor'].tolist(),
    'MAR': df_March['Valor'].tolist(),
    'ABR': df_April['Valor'].tolist(),
    'MAY': df_May['Valor'].tolist(),
    'JUN': df_June['Valor'].tolist(),
    'JUL': df_July['Valor'].tolist(),
    'AGO': df_August['Valor'].tolist(),
    'SEP': df_September['Valor'].tolist(),
    'OCT': df_October['Valor'].tolist(),
    'NOV': df_November['Valor'].tolist(),
    'DIC': df_December['Valor'].tolist()
}, index=range(len(YEAR9))) # El rango del nueva dataFrame esta dado por la longitud (cantidad de filas) de df de año
print(tb_mes_ano)
```

```python
#creamos la columna TOTAL para tener los datos por año
tb_mes_ano['TOTAL_AÑO'] = (tb_mes_ano.sum(axis=1)-tb_mes_ano['Año'])
tb_mes_ano['PROMEDIO_AÑO'] = tb_mes_ano['TOTAL_AÑO']/12
print(tb_mes_ano)
```

```python
tb_descripcion = tb_mes_ano.describe()
print(tb_descripcion)
```

```python
# Seleccionar filas donde esta la información de interes=> 0:Cantidad, 1: Promedio (mean), 3: minimo, 7: maximo

tb_resumen = tb_descripcion.iloc[[0,1,3,7]]
print(tb_resumen)
```

```python
# Obtenemos los dos parametros faltantes Total y Promedio

sum_resumn = pd.DataFrame(tb_mes_ano.sum())
sum_resumn.columns = ['Total']
sum_resumn = sum_resumn.T
print(sum_resumn)
```

```python
#Unimos los DataFrame

tb_fin = pd.concat([tb_resumen, sum_resumn])
print(tb_fin)
```
